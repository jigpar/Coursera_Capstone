```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
 
```

##### get the webpage content

```python
# get the RUL te
page_content = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
#print(type(page_content))
#print(page_content)

soup = BeautifulSoup(page_content, 'lxml')
#print(soup.prettify())   
```

##### Fectch just the table from the wiki page

```python
# fetch teable from soup
my_table = soup.find('table',{'class':'wikitable sortable'})
#print(my_table)
# find all the rows 
rows = my_table.findAll('tr')
#print(rows)
```

```python
# Creating a list of all the coluluns
Postcode = []
Borough = []
Neighbourhood = []
Postcode.clear()
Borough.clear()
Neighbourhood.clear()
print(len(Postcode), len(Borough), len(Neighbourhood))
```

```python
# loop throug all the rows, for each row find td, ignore th, 
#for each td, get td's text and append to the list
Postcode.clear()
Borough.clear()
Neighbourhood.clear()
print(len(Postcode), len(Borough), len(Neighbourhood))
prev_pc = ''
current_pc = ''
counter = 0
for row in rows:
    #line = row.text
    #print(line.split('\n'))
    tds = row.findAll('td')
    if (len(tds) == 3) and (tds[1].text != 'Not assigned'): #Only extract table body not heading
        #print(tds[0].text)
        #print(tds[1].text)
        #print(tds[2].text)
        #print(len(tds[2].text.strip()))
        #tdtext = td.text
        #print (len(tds))
        #print('adding:', Postcode.append(tds[0].text), Borough.append(tds[1].text), Neighbourhood.append(tds[2].text))
        """ old code
        Postcode.append(tds[0].text)
        Borough.append(tds[1].text)
        if(tds[2].text == 'Not assigned\n'):
            Neighbourhood.append(tds[1].text)
        else:
            Neighbourhood.append(tds[2].text.strip())
        """  
        current_pc = tds[0].text
        #print(tds[0].text, tds[1].text, tds[2].text)
        # if previous and current postcodes are not same, add to the list
        if (prev_pc != current_pc):
            Postcode.append(tds[0].text)
            Borough.append(tds[1].text)
            if(tds[2].text == 'Not assigned\n'):
                Neighbourhood.append(tds[1].text)
            else:
                Neighbourhood.append(tds[2].text.strip())
            counter += 1 #increment counter for next for next  postalcode
        # if two postcode are same, don't append to the list, instead of iterate back to previous rown and add Neighborhood comma separated
        elif (prev_pc == current_pc):
            #print ('counter: ', counter)
            #print('both pc same, prev and current', prev_pc, current_pc)
            #print(Neighbourhood[counter-1]) 
            Neighbourhood[counter-1] = Neighbourhood[counter-1] + ', ' + tds[2].text.strip()
            
        
        prev_pc = current_pc

#print(Neighbourhood)
```

```python
#create data frame from the Postcode, Borough and Neighbourhood data
df=pd.DataFrame()
df['Postcode'] = Postcode
df['Borough'] = Borough
df['Neighbourhood'] = Neighbourhood
df
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of New York City are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
# add markers to map
#Postcode	Borough	Neighbourhood	Latitude	Longitude
for lat, lng, borough, neighborhood, pc in zip(df_new['Latitude'], df_new['Longitude'], df_new['Borough'], df_new['Neighbourhood'], df_new['Postcode']):
    label = '{}, {}, {}'.format(neighborhood, borough, pc)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    

map_toronto
```
