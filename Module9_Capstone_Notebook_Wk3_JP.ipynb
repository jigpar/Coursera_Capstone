```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

##### get the webpage content

```python
# get the RUL te
page_content = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

soup = BeautifulSoup(page_content, 'lxml')
```

###### get the table from wiki page

```python
# fetch teable from soup
my_table = soup.find('table',{'class':'wikitable sortable'})
#print(my_table)
# find all the rows 
rows = my_table.findAll('tr')
```

##### creating all empty list to populate from the wiki table data

```python
# Creating a list of all the coluluns
Postalcode = []
Borough = []
Neighborhood = []
```

##### start the loop and populate the list baesd on assignment steps.

```python
# loop throug all the rows, for each row find td, ignore th, 
#for each td, get td's text and append to the list
Postalcode.clear()
Borough.clear()
Neighborhood.clear()
prev_pc = ''
current_pc = ''
counter = 0

for row in rows:
    tds = row.findAll('td')
    
    #Only extract table body not heading 
    if (len(tds) == 3) and (tds[1].text != 'Not assigned'): 
        #""" old code for reference
        #Postalcode.append(tds[0].text)
        #Borough.append(tds[1].text)
        #if(tds[2].text == 'Not assigned\n'):
        #   Neighborhood.append(tds[1].text)
        #else:
        #    Neighborhood.append(tds[2].text.strip())
        #"""  
        
        current_pc = tds[0].text
        
        if (prev_pc != current_pc):
            Postalcode.append(tds[0].text)
            Borough.append(tds[1].text)
            
            if(tds[2].text == 'Not assigned\n'):
                Neighborhood.append(tds[1].text)
            else:
                Neighborhood.append(tds[2].text.strip())
            counter += 1
            
        elif (prev_pc == current_pc):
            Neighborhood[counter-1] = Neighborhood[counter-1] + ', ' + tds[2].text.strip()
        
        prev_pc = current_pc

```

##### create dataframe and populate data

```python
#create data frame
df=pd.DataFrame()
df['Postalcode'] = Postalcode
df['Borough'] = Borough
df['Neighborhood'] = Neighborhood

# priting df.shape
print('df.shape', df.shape)
df
```

###### Now read the CSV file for latitude and longitude for postal codes and then merge two dataframes 
##### so we have a new dataframe with latitidue and longitude

```python
path='https://cocl.us/Geospatial_data/Geospatial_Coordinates.csv'
dfll = pd.read_csv(path)

dfll.rename(columns={'Postal Code': 'Postalcode'}, inplace=True)
df_merged = pd.merge(df, dfll, on ='Postalcode')
df_merged

#df_new2 = df.merge(dfll, left_on='Postcode', right_on='Postcode')
#df_new2

```

```python
#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

```

###### create Toronto Map as asked

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of New York City are {}, {}.'.format(latitude, longitude))
```

##### loopo through and  superimpose neighborhoods/boroughs on the map

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
# add markers to map
#Postcode	Borough	Neighbourhood	Latitude	Longitude
for lat, lng, borough, neighborhood, pc in zip(df_merged['Latitude'], df_merged['Longitude'], df_merged['Borough'], df_merged['Neighborhood'], df_merged['Postalcode']):
    label = '{}, {}, {}'.format(neighborhood, borough, pc)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    

map_toronto
```
